```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from google.colab import files
files.upload()
```

```python
df = pd.read_csv('COCAINE_DEALING.csv')
df
```

```python
graph = {}
for index, row in df.iterrows():
    source = row['Unnamed: 0']

    if source not in graph:
        graph[source] = []

    for target_col, value in row.drop('Unnamed: 0').items():
        if value > 0:
            target = target_col
            if target not in graph:
                graph[target] = []
            graph[source].append(target)

print("Graph:", graph)

graph_dict = graph
graph = nx.Graph(graph_dict)
print("+")
print("Nodes:", graph.number_of_nodes())
print("Edges:", graph.number_of_edges())

dc = nx.degree_centrality(graph)
print("+")
print("Degree Centrality:", dc)

bc = nx.betweenness_centrality(graph)
print("+")
print("Betweenness Centrality:", bc)
```

```python
pos = nx.spring_layout(graph)
plt.figure(figsize=(20, 8))
nx.draw_networkx(graph, pos=pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()

density = nx.density(graph)
print("\n")
print("Density:", density)

cluster = nx.average_clustering(graph)
print("+")
print("Average Cluster:", cluster)
```

```python
newgraph = {}
for index, row in dropkcol.iterrows():
    source = row['Unnamed: 0']

    if source not in newgraph:
        newgraph[source] = []

    for target_col, value in row.drop('Unnamed: 0').items():
        if value > 0:
            target = target_col
            if target not in newgraph:
                newgraph[target] = []
            newgraph[source].append(target)

print("Graph:", newgraph)

newgraph_dict = newgraph
newgraph_nx = nx.Graph(newgraph_dict)
print("+")
print("Nodes:", newgraph_nx.number_of_nodes())
print("Edges:", newgraph_nx.number_of_edges())

newdc = nx.degree_centrality(newgraph_nx)
print("+")
print("Degree Centrality:", newdc)

newbc = nx.betweenness_centrality(newgraph_nx)
print("+")
print("Betweenness Centrality:", newbc)
```

```python
pos = nx.spring_layout(newgraph_nx)
plt.figure(figsize=(20, 8))
nx.draw_networkx(newgraph_nx, pos=pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()

newdensity = nx.density(newgraph_nx)
print("\n")
print("Density:", newdensity)

newcluster = nx.average_clustering(newgraph_nx)
print("+")
print("Average Cluster:", newcluster)
```

# Observations

The data analyzed above was the [Cocaine Dealing Natarajan](https://sites.google.com/site/ucinetsoftware/datasets/covert-networks/cocaine-dealing-natarajan) set. I chose to analyze this set after narrowing down which sets had the smallest amounts of data that were *not* stacked.\
\
To find the weakest link, I sought out the center of all connections by sorting values by *betweenness centrality*, which showed that **Kay** had the highest value. I removed Kay by dropping their column and row. \
\
Before removing Kay, the data had a density value of 0.106. After removal, the density was valued at 0.046. This means it was almost reduced in half.\
\
Visually, it is evident that removing Kay completely severed multiple connections, leaving many nodes on their own.
